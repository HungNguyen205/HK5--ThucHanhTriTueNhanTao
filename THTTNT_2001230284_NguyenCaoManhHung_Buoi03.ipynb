<div style="text-align: justify;">
&emsp;&emsp;NhÃ³m hÃ m nÃ y chá»‹u trÃ¡ch nhiá»‡m khá»Ÿi táº¡o bÃ n cá», xÃ¡c Ä‘á»‹nh lÆ°á»£t Ä‘i, tÃ¬m cÃ¡c nÆ°á»›c Ä‘i há»£p lá»‡ vÃ  cáº­p nháº­t bÃ n cá» sau khi Ä‘i.


```python
# HÃ m khá»Ÿi táº¡o tráº¡ng thÃ¡i bÃ n cá» ban Ä‘áº§u
def initial_state(n):
    """
    Khá»Ÿi táº¡o vÃ  tráº£ vá» bÃ n cá» trá»‘ng kÃ­ch thÆ°á»›c n x n.
    """
    return [[EMPTY for _ in range(n)] for _ in range(n)]

# HÃ m xÃ¡c Ä‘á»‹nh lÆ°á»£t ngÆ°á»i chÆ¡i hiá»‡n táº¡i
def player(board):
    """
    Tráº£ vá» ngÆ°á»i chÆ¡i cÃ³ lÆ°á»£t Ä‘i tiáº¿p theo dá»±a trÃªn tráº¡ng thÃ¡i bÃ n cá».
    - Äáº¿m sá»‘ quÃ¢n cá» Ä‘Ã£ Ä‘Ã¡nh Ä‘á»ƒ xÃ¡c Ä‘á»‹nh lÆ°á»£t (cháºµn lÃ  ngÆ°á»i chÆ¡i, láº» lÃ  mÃ¡y).
    """
    count = 0
    for i in board:
        for j in i:
            if j: # Náº¿u Ã´ khÃ´ng pháº£i lÃ  None (Ä‘Ã£ cÃ³ ngÆ°á»i Ä‘Ã¡nh)
                count += 1

    # Náº¿u sá»‘ nÆ°á»›c Ä‘i lÃ  láº» -> LÆ°á»£t cá»§a mÃ¡y (ngÆ°á»i Ä‘i sau)
    if count % 2 != 0:
        return ai
    # Náº¿u sá»‘ nÆ°á»›c Ä‘i lÃ  cháºµn -> LÆ°á»£t cá»§a ngÆ°á»i chÆ¡i (ngÆ°á»i Ä‘i trÆ°á»›c)
    return user

# HÃ m láº¥y danh sÃ¡ch cÃ¡c nÆ°á»›c Ä‘i há»£p lá»‡
def actions(board):
    """
    Tráº£ vá» táº­p há»£p táº¥t cáº£ cÃ¡c nÆ°á»›c Ä‘i kháº£ thi (i, j) trÃªn bÃ n cá».
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

# HÃ m tráº£ vá» tráº¡ng thÃ¡i bÃ n cá» má»›i (Transition Model)
def result(board, action):
    """
    Tráº£ vá» tráº¡ng thÃ¡i bÃ n cá» Má»šI sau khi thá»±c hiá»‡n nÆ°á»›c Ä‘i (i, j).
    """
    curr_player = player(board)       # XÃ¡c Ä‘á»‹nh xem ai Ä‘ang Ä‘i nÆ°á»›c nÃ y
    result_board = copy.deepcopy(board) # Táº¡o báº£n sao bÃ n cá»
    (i, j) = action
    result_board[i][j] = curr_player  # ÄÃ¡nh dáº¥u nÆ°á»›c Ä‘i
    return result_board
```

###3. CÃ¡c hÃ m kiá»ƒm tra káº¿t quáº£ vÃ  luáº­t chÆ¡i
<div style="text-align: justify;">
&emsp;&emsp;NhÃ³m hÃ m nÃ y kiá»ƒm tra xem ai tháº¯ng (ngang, dá»c, chÃ©o), game Ä‘Ã£ káº¿t thÃºc chÆ°a (terminal) vÃ  cháº¥m Ä‘iá»ƒm bÃ n cá» (utility).


```python
# HÃ m kiá»ƒm tra tháº¯ng theo hÃ ng ngang
def get_horizontal_winner(board):
    """
    Kiá»ƒm tra xem cÃ³ ngÆ°á»i tháº¯ng theo cÃ¡c hÃ ng ngang hay khÃ´ng.
    Tráº£ vá» kÃ½ tá»± ngÆ°á»i tháº¯ng ('X' hoáº·c 'O') hoáº·c None.
    """
    winner_val = None
    board_len = len(board)

    for i in range(board_len):
        winner_val = board[i][0] # Láº¥y Ã´ Ä‘áº§u tiÃªn cá»§a hÃ ng lÃ m máº«u
        for j in range(board_len):
            if board[i][j] != winner_val: # Náº¿u cÃ³ Ã´ khÃ¡c vá»›i Ã´ máº«u -> HÃ ng nÃ y khÃ´ng tháº¯ng
                winner_val = None
        if winner_val: # Náº¿u winner_val váº«n khÃ¡c None vÃ  khÃ¡c False -> TÃ¬m tháº¥y ngÆ°á»i tháº¯ng
            return winner_val
    return winner_val

# HÃ m kiá»ƒm tra tháº¯ng theo hÃ ng dá»c
def get_vertical_winner(board):
    """
    Kiá»ƒm tra xem cÃ³ ngÆ°á»i tháº¯ng theo cÃ¡c cá»™t dá»c hay khÃ´ng.
    """
    winner_val = None
    board_len = len(board)

    for i in range(board_len):
        winner_val = board[0][i] # Láº¥y Ã´ Ä‘áº§u tiÃªn cá»§a cá»™t lÃ m máº«u
        for j in range(board_len):
            if board[j][i] != winner_val: # So sÃ¡nh cÃ¡c Ã´ trong cá»™t Ä‘Ã³
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val

# HÃ m kiá»ƒm tra tháº¯ng theo Ä‘Æ°á»ng chÃ©o
def get_diagonal_winner(board):
    """
    Kiá»ƒm tra xem cÃ³ ngÆ°á»i tháº¯ng theo hai Ä‘Æ°á»ng chÃ©o (chÃ­nh vÃ  phá»¥).
    """
    winner_val = None
    board_len = len(board)

    # 1. Kiá»ƒm tra Ä‘Æ°á»ng chÃ©o chÃ­nh
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
    if winner_val:
        return winner_val

    # 2. Kiá»ƒm tra Ä‘Æ°á»ng chÃ©o phá»¥
    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None
    return winner_val

# HÃ m xÃ¡c Ä‘á»‹nh ngÆ°á»i tháº¯ng chung cuá»™c
def winner(board):
    """
    Tráº£ vá» ngÆ°á»i tháº¯ng cuá»™c cá»§a trÃ² chÆ¡i (náº¿u cÃ³).
    Káº¿t há»£p káº¿t quáº£ tá»« kiá»ƒm tra Ngang, Dá»c vÃ  ChÃ©o.
    """
    # Náº¿u tÃ¬m tháº¥y ngÆ°á»i tháº¯ng á»Ÿ báº¥t ká»³ hÆ°á»›ng nÃ o, tráº£ vá» ngÆ°á»i Ä‘Ã³. NgÆ°á»£c láº¡i tráº£ vá» None.
    winner_val = get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board) or None
    return winner_val

# HÃ m kiá»ƒm tra tráº¡ng thÃ¡i káº¿t thÃºc
def terminal(board):
    """
    Tráº£ vá» True náº¿u trÃ² chÆ¡i Ä‘Ã£ káº¿t thÃºc, False náº¿u chÆ°a.
    TrÃ² chÆ¡i káº¿t thÃºc khi:
    1. ÄÃ£ cÃ³ ngÆ°á»i tháº¯ng (winner != None).
    2. Hoáº·c bÃ n cá» Ä‘Ã£ Ä‘áº§y (HÃ²a).
    """
    # TrÆ°á»ng há»£p 1: CÃ³ ngÆ°á»i tháº¯ng
    if winner(board) != None:
        return True

    # TrÆ°á»ng há»£p 2: Kiá»ƒm tra bÃ n cá» cÃ²n Ã´ trá»‘ng khÃ´ng
    for i in board:
        for j in i:
            if j == EMPTY: # Náº¿u cÃ²n Ã­t nháº¥t 1 Ã´ trá»‘ng
                return False

    # Náº¿u khÃ´ng cÃ²n Ã´ trá»‘ng nÃ o -> HÃ²a -> Káº¿t thÃºc
    return True

# HÃ m Ä‘á»‹nh giÃ¡
def utility(board):
    """
    Tráº£ vá» giÃ¡ trá»‹ lá»£i Ã­ch cá»§a tráº¡ng thÃ¡i káº¿t thÃºc.
    - Tráº£ vá» 1 náº¿u X tháº¯ng.
    - Tráº£ vá» -1 náº¿u O tháº¯ng.
    - Tráº£ vá» 0 náº¿u HÃ²a.
    """
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0
```

###4. Thuáº­t toÃ¡n Minimax
<div style="text-align: justify;">
&emsp;&emsp;Sá»­ dá»¥ng Ä‘á»‡ quy (maxValue vÃ  minValue) Ä‘á»ƒ tÃ¬m nÆ°á»›c Ä‘i tá»‘i Æ°u nháº¥t.</div>


```python
# HÃ m tÃ¬m giÃ¡ trá»‹ lá»›n nháº¥t
def maxValue(state):
    """
    Tráº£ vá» giÃ¡ trá»‹ tiá»‡n Ã­ch lá»›n nháº¥t cÃ³ thá»ƒ Ä‘áº¡t Ä‘Æ°á»£c tá»« tráº¡ng thÃ¡i nÃ y.
    """
    # 1. Äiá»u kiá»‡n dá»«ng: Náº¿u game káº¿t thÃºc, tráº£ vá» Ä‘iá»ƒm sá»‘
    if terminal(state):
        return utility(state)

    # 2. Khá»Ÿi táº¡o giÃ¡ trá»‹ ban Ä‘áº§u lÃ  Ã¢m vÃ´ cÃ¹ng (tá»‡ nháº¥t vá»›i MAX)
    v = -math.inf

    # 3. Duyá»‡t qua cÃ¡c hÃ nh Ä‘á»™ng, tÃ¬m giÃ¡ trá»‹ lá»›n nháº¥t tá»« cÃ¡c pháº£n á»©ng cá»§a MIN
    for action in actions(state):
        v = max(v, minValue(result(state, action)))

    return v

# HÃ m tÃ¬m giÃ¡ trá»‹ nhá» nháº¥t
def minValue(state):
    """
    Tráº£ vá» giÃ¡ trá»‹ tiá»‡n Ã­ch nhá» nháº¥t cÃ³ thá»ƒ Ä‘áº¡t Ä‘Æ°á»£c tá»« tráº¡ng thÃ¡i nÃ y.
    """
    # 1. Äiá»u kiá»‡n dá»«ng
    if terminal(state):
        return utility(state)

    # 2. Khá»Ÿi táº¡o giÃ¡ trá»‹ ban Ä‘áº§u lÃ  dÆ°Æ¡ng vÃ´ cÃ¹ng (tá»‡ nháº¥t vá»›i MIN)
    v = math.inf

    # 3. Duyá»‡t qua cÃ¡c hÃ nh Ä‘á»™ng, tÃ¬m giÃ¡ trá»‹ nhá» nháº¥t tá»« cÃ¡c pháº£n á»©ng cá»§a MAX
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))

    return v

# HÃ m chÃ­nh: Quyáº¿t Ä‘á»‹nh nÆ°á»›c Ä‘i tá»‘i Æ°u
def minimax(board):
    """
    Tráº£ vá» nÆ°á»›c Ä‘i tá»‘i Æ°u nháº¥t cho ngÆ°á»i chÆ¡i hiá»‡n táº¡i.
    """
    current_player = player(board)

    # --- TRÆ¯á»œNG Há»¢P 1: LÆ°á»£t cá»§a MAX ---
    if current_player == X:
        best_val = -math.inf
        best_move = None

        # Duyá»‡t táº¥t cáº£ cÃ¡c nÆ°á»›c Ä‘i cÃ³ thá»ƒ
        for action in actions(board):
            # Giáº£ Ä‘á»‹nh Ä‘i nÆ°á»›c nÃ y, rá»“i xem MIN sáº½ lÃ m gÃ¬ (gá»i minValue)
            val = minValue(result(board, action))

            # Náº¿u tÃ¬m tháº¥y nÆ°á»›c Ä‘i cÃ³ giÃ¡ trá»‹ cao hÆ¡n giÃ¡ trá»‹ tá»‘t nháº¥t hiá»‡n táº¡i
            if val > best_val:
                best_val = val
                best_move = action

        return best_move

    # --- TRÆ¯á»œNG Há»¢P 2: LÆ°á»£t cá»§a MIN ---
    else:
        best_val = math.inf
        best_move = None

        # Duyá»‡t táº¥t cáº£ cÃ¡c nÆ°á»›c Ä‘i cÃ³ thá»ƒ
        for action in actions(board):
            # Giáº£ Ä‘á»‹nh Ä‘i nÆ°á»›c nÃ y, rá»“i xem MAX sáº½ lÃ m gÃ¬ (gá»i maxValue)
            val = maxValue(result(board, action))

            # Náº¿u tÃ¬m tháº¥y nÆ°á»›c Ä‘i cÃ³ giÃ¡ trá»‹ tháº¥p hÆ¡n (tá»‘t cho MIN)
            if val < best_val:
                best_val = val
                best_move = action

        return best_move
```

#ALPHA-BETA PRUNING

###1. Khá»Ÿi táº¡o vÃ  CÃ¡c hÃ m tiá»‡n Ã­ch

```python
import math
import copy

# X, O, EMPTY, user, ai Ä‘Æ°á»£c giáº£ Ä‘á»‹nh lÃ  Ä‘Ã£ Ä‘Æ°á»£c khai bÃ¡o toÃ n cá»¥c bá»Ÿi script chÃ­nh vÃ  cÃ¡c Ã´ trÆ°á»›c Ä‘Ã³.
# HÃ m 'initial_state' cÅ©ng Ä‘Æ°á»£c Ä‘á»‹nh nghÄ©a trong má»™t Ã´ trÆ°á»›c Ä‘Ã³ vÃ  Ä‘Æ°á»£c sá»­ dá»¥ng Ä‘á»ƒ khá»Ÿi táº¡o bÃ n cá».
# CÃ¡c hÃ m 'PrintBoard' vÃ  'GetAvailableCells' trong pháº§n Alpha-Beta gá»‘c Ä‘Æ°á»£c thiáº¿t káº¿ cho
# Biá»ƒu diá»…n bÃ n cá» 1D vÃ  khÃ´ng tÆ°Æ¡ng thÃ­ch vá»›i bÃ n cá» 2D Ä‘Æ°á»£c sá»­ dá»¥ng á»Ÿ Ä‘Ã¢y.
# ChÃºng ta sáº½ dá»±a vÃ o 'actions(board)' tá»« pháº§n Minimax Ä‘á»ƒ láº¥y cÃ¡c nÆ°á»›c Ä‘i kháº£ thi vÃ  dÃ¹ng lá»‡nh in chuáº©n cá»§a Python cho bÃ n cá».
```

###2. HÃ m Ä‘á»‹nh giÃ¡ vÃ  Kiá»ƒm tra káº¿t thÃºc
<div style="text-align: justify;">
&emsp;&emsp;XÃ¡c Ä‘á»‹nh tráº¡ng thÃ¡i tháº¯ng/thua/hÃ²a cá»§a bÃ n cá» dá»±a trÃªn luáº­t chÆ¡i (ngang, dá»c, chÃ©o)


```python
# HÃ m 'GetWinner' trong pháº§n Alpha-Beta gá»‘c Ä‘Æ°á»£c thiáº¿t káº¿ cho biá»ƒu diá»…n bÃ n cá» 1D.
# ChÃºng ta sáº½ dá»±a vÃ o hÃ m 'winner(board)' Ä‘Æ°á»£c Ä‘á»‹nh nghÄ©a trong pháº§n Minimax
# HÃ m nÃ y tÆ°Æ¡ng thÃ­ch vá»›i bÃ n cá» 2D Ä‘Æ°á»£c sá»­ dá»¥ng trong chÆ°Æ¡ng trÃ¬nh chÃ­nh.
```

###3. Thuáº­t toÃ¡n Alpha-Beta
<div style="text-align: justify;">
&emsp;&emsp;HÃ m Ä‘á»‡ quy thá»±c hiá»‡n viá»‡c duyá»‡t cÃ¢y, tÃ­nh Ä‘iá»ƒm vÃ  cáº¯t tá»‰a nhÃ¡nh thá»«a</div>


```python
# HÃ m nÃ y thá»±c hiá»‡n logic cáº¯t tá»‰a Alpha-Beta Ä‘á»‡ quy.
# NÃ³ giáº£ Ä‘á»‹nh ráº±ng X, O, EMPTY, terminal, utility, actions, vÃ  result cÃ³ thá»ƒ truy cáº­p Ä‘Æ°á»£c tá»« cÃ¡c Ã´ trÆ°á»›c Ä‘Ã³.

def _alpha_beta_recursive(board, depth, alpha, beta, is_max_player):
    # TrÆ°á»ng há»£p cÆ¡ sá»Ÿ: náº¿u trÃ² chÆ¡i káº¿t thÃºc, tráº£ vá» tiá»‡n Ã­ch (Ä‘iá»ƒm sá»‘) cá»§a tráº¡ng thÃ¡i.
    # TÃ¹y chá»n, bao gá»“m Ä‘á»™ sÃ¢u trong tiá»‡n Ã­ch Ä‘á»ƒ Æ°u tiÃªn tháº¯ng nhanh hÆ¡n hoáº·c thua cháº­m hÆ¡n.
    if terminal(board):
        score = utility(board)
        if score == 1:  # NgÆ°á»i chÆ¡i MAX tháº¯ng
            return score + (10 - depth) # Æ¯u tiÃªn tháº¯ng nhanh hÆ¡n
        elif score == -1: # NgÆ°á»i chÆ¡i MIN tháº¯ng
            return score - (10 - depth) # Æ¯u tiÃªn lÃ m cho MAX thua cháº­m hÆ¡n
        else: # HÃ²a
            return score

    # Náº¿u Ä‘áº¿n lÆ°á»£t ngÆ°á»i chÆ¡i MAX (vÃ­ dá»¥: 'X')
    if is_max_player:
        max_eval = -math.inf
        # Láº·p qua táº¥t cáº£ cÃ¡c nÆ°á»›c Ä‘i cÃ³ thá»ƒ báº±ng cÃ¡ch sá»­ dá»¥ng hÃ m 'actions' cho bÃ n cá» 2D.
        for action in actions(board):
            new_board = result(board, action) # Láº¥y tráº¡ng thÃ¡i bÃ n cá» má»›i sau khi thá»±c hiá»‡n hÃ nh Ä‘á»™ng.
            # Gá»i Ä‘á»‡ quy cho ngÆ°á»i chÆ¡i MIN (is_max_player=False).
            eval = _alpha_beta_recursive(new_board, depth + 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval) # Cáº­p nháº­t alpha (giÃ¡ trá»‹ tá»‘t nháº¥t tÃ¬m tháº¥y cho MAX cho Ä‘áº¿n nay)
            if beta <= alpha: # Äiá»u kiá»‡n cáº¯t tá»‰a Alpha-Beta: Náº¿u nÆ°á»›c Ä‘i hiá»‡n táº¡i cá»§a MAX Ä‘Ã£ tá»‡ hÆ¡n lá»±a chá»n tá»‘t nháº¥t cá»§a MIN, hÃ£y cáº¯t tá»‰a.
                break
        return max_eval
    # Náº¿u Ä‘áº¿n lÆ°á»£t ngÆ°á»i chÆ¡i MIN (vÃ­ dá»¥: 'O')
    else: # is_max_player lÃ  False
        min_eval = math.inf
        # Láº·p qua táº¥t cáº£ cÃ¡c nÆ°á»›c Ä‘i cÃ³ thá»ƒ báº±ng cÃ¡ch sá»­ dá»¥ng hÃ m 'actions' cho bÃ n cá» 2D.
        for action in actions(board):
            new_board = result(board, action) # Láº¥y tráº¡ng thÃ¡i bÃ n cá» má»›i sau khi thá»±c hiá»‡n hÃ nh Ä‘á»™ng.
            # Gá»i Ä‘á»‡ quy cho ngÆ°á»i chÆ¡i MAX (is_max_player=True).
            eval = _alpha_beta_recursive(new_board, depth + 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval) # Cáº­p nháº­t beta (giÃ¡ trá»‹ tá»‘t nháº¥t tÃ¬m tháº¥y cho MIN cho Ä‘áº¿n nay)
            if beta <= alpha: # Äiá»u kiá»‡n cáº¯t tá»‰a Alpha-Beta: Náº¿u nÆ°á»›c Ä‘i hiá»‡n táº¡i cá»§a MIN Ä‘Ã£ tá»‘t hÆ¡n lá»±a chá»n tá»‘t nháº¥t cá»§a MAX, hÃ£y cáº¯t tá»‰a.
                break
        return min_eval
```

```python
# HÃ m nÃ y xÃ¡c Ä‘á»‹nh nÆ°á»›c Ä‘i tá»‘t nháº¥t cho ngÆ°á»i chÆ¡i AI báº±ng cÃ¡ch sá»­ dá»¥ng Cáº¯t tá»‰a Alpha-Beta.
# NÃ³ giáº£ Ä‘á»‹nh X, O, player, actions, result, vÃ  _alpha_beta_recursive cÃ³ thá»ƒ truy cáº­p Ä‘Æ°á»£c.

def FindBestMove(board, ai_player_char):
    """
    TÃ¬m nÆ°á»›c Ä‘i tá»‘i Æ°u nháº¥t cho ngÆ°á»i chÆ¡i AI báº±ng Alpha-Beta Pruning.
    Sá»­ dá»¥ng board 2D vÃ  cÃ¡c hÃ m game logic tá»« pháº§n Minimax.
    """
    best_move = None
    current_ai_player = ai_player_char

    # Náº¿u AI lÃ  ngÆ°á»i chÆ¡i MAX (vÃ­ dá»¥: 'X')
    if current_ai_player == X:
        best_val = -math.inf
        # Láº·p qua táº¥t cáº£ cÃ¡c hÃ nh Ä‘á»™ng cÃ³ thá»ƒ
        for action in actions(board):
            new_board = result(board, action) # MÃ´ phá»ng AI thá»±c hiá»‡n nÆ°á»›c Ä‘i nÃ y
            # Gá»i hÃ m alpha-beta Ä‘á»‡ quy, lÆ°á»£t tiáº¿p theo sáº½ lÃ  cá»§a ngÆ°á»i chÆ¡i MIN (Ä‘á»‘i thá»§)
            move_val = _alpha_beta_recursive(new_board, 0, -math.inf, math.inf, False)
            if move_val > best_val:
                best_val = move_val
                best_move = action
    # Náº¿u AI lÃ  ngÆ°á»i chÆ¡i MIN (vÃ­ dá»¥: 'O')
    else: # current_ai_player == O
        best_val = math.inf
        # Láº·p qua táº¥t cáº£ cÃ¡c hÃ nh Ä‘á»™ng cÃ³ thá»ƒ
        for action in actions(board):
            new_board = result(board, action) # MÃ´ phá»ng AI thá»±c hiá»‡n nÆ°á»›c Ä‘i nÃ y
            # Gá»i hÃ m alpha-beta Ä‘á»‡ quy, lÆ°á»£t tiáº¿p theo sáº½ lÃ  cá»§a ngÆ°á»i chÆ¡i MAX (Ä‘á»‘i thá»§)
            move_val = _alpha_beta_recursive(new_board, 0, -math.inf, math.inf, True)
            if move_val < best_val:
                best_val = move_val
                best_move = action

    return best_move
```
